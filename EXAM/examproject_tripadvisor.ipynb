```python
from bs4 import BeautifulSoup
import requests
import re
import scraping_class
logfile = 'log.csv'
connector = scraping_class.Connector(logfile)
```

```python
# tripadvisor, get all url pages

numbers = [30*i for i in range(49)]
```

```python
# Tripadvisor stuff

list_of_rest_links = []

for number in numbers:
    url = f'https://www.tripadvisor.dk/RestaurantSearch-g189541-oa{number}-a_date.2020__2D__08__2D__21-a_people.2-a_time.20%3A00%3A00-a_zur.2020__5F__08__5F__21-Copenhag.html#EATERY_LIST_CONTENTS'
    response = connector.get(url,'url_getter')
    html = response[0].text
    soup = BeautifulSoup(html,'html.parser')
    list_of_rest_links.append([link['href'] for link in soup.find_all(class_ = '_15_ydu6b',href = True)])# = [link['href'] for link in soup.find_all(class_ = '_15_ydu6b',href = True)]

list_of_rest_links = [item for sublist in list_of_rest_links for item in sublist]

tripadvisor = 'https://www.tripadvisor.dk'

list_of_rest_links = [tripadvisor+list_of_rest_links[i] for list_of_rest_links[i] in list_of_rest_links]

list_of_rest_links

#for link in soup.find_all(class_ = 'reviewlink', href = True):
#    print(link['href'])

#list_of_game_links
```

```python
# Test to get stuff

response2 = connector.get(list_of_rest_links[0],'rest_data_getter')

html2 = response2[0].text

soup2 = BeautifulSoup(html2,'html.parser')

rest_name = soup2.find(class_ = '_3a1XQ88S').get_text(strip=True)
rest_location = soup2.find(class_ = '_2saB_OSe').get_text(strip=True)
rest_rating = soup2.find(class_ = 'r2Cf69qf').get_text(strip=True).replace(",", ".")
rest_cuisine = str(soup2.find_all(class_ = '_1XLfiSsv')).split(",")[1][24:]
rest_no_review = soup2.find(class_ = '_10Iv7dOs').get_text(strip=True)[:-12]
rest_price = soup2.find(class_ = '_1XLfiSsv').get_text(strip=True)
rest_price_catdollar = soup2.find(class_ = '_2mn01bsa').get_text(strip=True).count('$')
rest_price_catdash = soup2.find(class_ = '_2mn01bsa').get_text(strip=True).count('-')
rest_price_cat = rest_price_catdollar / (1 + rest_price_catdash)
```
