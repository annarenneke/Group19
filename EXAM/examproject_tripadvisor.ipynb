```python
from bs4 import BeautifulSoup
import requests
import re
import time
import scraping_class
import pandas as pd
import numpy as np
logfile = 'log.csv'
connector = scraping_class.Connector(logfile)
```

```python
# tripadvisor, get all url pages

numbers = [30*i for i in range(49)]
```

```python
# Tripadvisor stuff

list_of_rest_links_pre = []

timer = 0

# The following loop takes approx 1.5 minutes.

for number in numbers:
    
    start = time.time()
    
    url = f'https://www.tripadvisor.dk/RestaurantSearch-g189541-oa{number}-a_date.2020__2D__08__2D__21-a_people.2-a_time.20%3A00%3A00-a_zur.2020__5F__08__5F__21-Copenhag.html#EATERY_LIST_CONTENTS'
    response = connector.get(url,'rest_url_getter')
    html = response[0].text
    soup = BeautifulSoup(html,'html.parser')
    list_of_rest_links_pre.append([link['href'] for link in soup.find_all(class_ = '_15_ydu6b',href = True)])# = [link['href'] for link in soup.find_all(class_ = '_15_ydu6b',href = True)]
    end = time.time()
    timer = timer + end - start
    timer_accu = str(round(timer/60, 1))
    print("The loop has run for " + timer_accu + " minutes." )
```

```python
list_of_rest_links_01 = [item for sublist in list_of_rest_links_pre for item in sublist]

list_of_rest_links_01[:5]
```

```python
tripadvisor = 'https://www.tripadvisor.dk'

list_of_rest_links = [tripadvisor + i for i in list_of_rest_links_01]

del list_of_rest_links_01, list_of_rest_links_pre, html

list_of_rest_links[:10]
```

```python
rest_name = []
rest_location = []
rest_rating = []
rest_cuisine = []
rest_cuisine_2 = []
rest_cuisine_3 = []
rest_no_review = []
rest_price_cat = []

timer = 0

# The following loop takes approx 64 minutes.

for link in list_of_rest_links:
    
    start = time.time()
    
    response = connector.get(link,'rest_data_getter')
    html = response[0].text
    soup = BeautifulSoup(html,'html.parser')
    try:
        rest_name1 = soup.find(class_ = '_3a1XQ88S').get_text(strip=True)
    except:
        rest_name1 = ''
    try:
        rest_location1 = soup.find(class_ = '_2saB_OSe').get_text(strip=True)
    except:
        rest_location1 = ''
    try:
        rest_rating1 = soup.find(class_ = 'r2Cf69qf').get_text(strip=True).replace(",", ".")
    except:
        rest_rating1 = ''
    try:
        rest_cuisine1 = str(soup.find_all(class_ = '_2mn01bsa')).split(",")[1].split('html">')[1].replace('</a>', '')
    except:
        rest_cuisine1 = ''
    try:
        rest_cuisine2 = str(soup.find_all(class_ = '_2mn01bsa')).split(",")[2].split('html">')[1].replace('</a>', '')
    except:
        rest_cuisine2 = ''
    try:
        rest_cuisine3 = str(soup.find_all(class_ = '_2mn01bsa')).split(",")[3].split('html">')[1].replace('</a>]', '')
    except:
        rest_cuisine3 = ''
    try:
        rest_no_review1 = soup.find(class_ = '_10Iv7dOs').get_text(strip=True).split(" ")[0].replace('.', '')
    except:
        rest_no_review1 = ''
    try:
        rest_price_catdollar1 = soup.find(class_ = '_2mn01bsa').get_text(strip=True).count('$')
    except:
        rest_price_catdollar1 = ''
    try:
        rest_price_catdash1 = soup.find(class_ = '_2mn01bsa').get_text(strip=True).count('-')
    except:
        rest_price_catdash1 = ''
    try:
        rest_price_cat1 = rest_price_catdollar1 / (1 + rest_price_catdash1)
    except:
        rest_price_cat1 = ''
    rest_name.append(rest_name1)
    rest_location.append(rest_location1)
    rest_rating.append(rest_rating1)
    rest_cuisine.append(rest_cuisine1)
    rest_cuisine_2.append(rest_cuisine2)
    rest_cuisine_3.append(rest_cuisine3)
    rest_no_review.append(rest_no_review1)
    rest_price_cat.append(rest_price_cat1)
    
    end = time.time()
    timer = timer + end - start
    timer_accu = str(round(timer/60, 1))
    
    #time.sleep(0.5)
    
    print("Completed " + str(len(rest_name)) + " of 1807 links in " + timer_accu + " minutes.")
```

```python
df = pd.DataFrame({'links': list_of_rest_links, 'rest_name' : rest_name, 'rest_location' : rest_location, 'rest_rating' : rest_rating, 'rest_cuisine' : rest_cuisine, 'rest_cuisine_2' : rest_cuisine_2, 'rest_cuisine_3' : rest_cuisine_3, 'rest_no_review' : rest_no_review, 'rest_price_cat' : rest_price_cat})

df_overlinksdata = pd.DataFrame({'links': list_of_rest_links, 'rest_name' : rest_name, 'rest_location' : rest_location, 'rest_rating' : rest_rating, 'rest_cuisine' : rest_cuisine, 'rest_cuisine_2' : rest_cuisine_2, 'rest_cuisine_3' : rest_cuisine_3, 'rest_no_review' : rest_no_review, 'rest_price_cat' : rest_price_cat})
df_overlinksdata.to_csv('overlinksdata.csv', index=False)

df2 = df.drop_duplicates(subset=['rest_name'], keep='first').copy()

df2
```

```python
links_rest = pd.read_csv('links_rest.csv',sep=',')
links_rest = list(links_rest['links'])
links_rest[:10]
```

```python
# Following loop runs for approx. 39 minutes.

danish_revs = []

timer = 0

for link in links_rest:

    start = time.time()

    url = link
    response = connector.get(url,'danish_revs_getter')
    html = response[0].text
    soup = BeautifulSoup(html,'html.parser')

    try:
        number_of_revs = int(str(soup.find(class_ = 'count')).replace('<span class="count">(', '').replace(')</span>', '').replace('.', ''))
    except:
        number_of_revs = np.nan
    
    danish_revs.append(number_of_revs)
    end = time.time()
    timer = timer + end - start
    timer_accu = str(round(timer/60, 1))
    print("Completed " + str(len(danish_revs)) + " of 1294 links after " + timer_accu + " minutes." )
```

```python
df_danish_revs = pd.DataFrame(danish_revs, columns=["danish_revs"])
df_danish_revs.to_csv('danish_revs.csv', index=False)
```

```python
last_link_no = [(-(-int(pages) // 10) -1) for pages in danish_revs]

last_link_no[:20]
```

```python
i = 0
url_test = []
j = 0

timer = 0

# While loop lasts for approx. 5.3 seconds.

while i <= last_link_no[j] and j < len(last_link_no) - 1:
    
    start = time.time()
    
    line = links_rest[j]
    index = line.find('Reviews-')
    sub_url = line[:index] + 'Reviews-or' + str(i) + '0' + line[index:][7:]
    
    print(sub_url)
    print(line)
    print(i)
    print(j)
    
    url_test.append(sub_url)
    print(len(url_test))
    
    end = time.time()
    timer = timer + end - start
    timer_accu = str(round(timer/1, 1))
    print("The while loop has run for " + timer_accu + " seconds." )
    
    if i == last_link_no[j]:
        j += 1
        i = 0
    else:
        i += 1
```

```python
# Getting all individual review scores.

rest_rates = []

rest_rating_date = []

timer = 0

sub_rest_name = []

for link in url_test[:259]:
    
    start = time.time()
    
    url = link
    response = connector.get(url,'ind_rating_getter')
    html = response[0].text
    soup = BeautifulSoup(html,'html.parser')
    
    ind_reviewtest = soup.find_all(class_ = 'ui_column is-9')#.find_all(class_ = 'ui_bubble_rating bubble_50')

    rate5 = [n.find_all(class_ = 'ui_bubble_rating bubble_50') for n in ind_reviewtest]
    rate4 = [n.find_all(class_ = 'ui_bubble_rating bubble_40') for n in ind_reviewtest]
    rate3 = [n.find_all(class_ = 'ui_bubble_rating bubble_30') for n in ind_reviewtest]
    rate2 = [n.find_all(class_ = 'ui_bubble_rating bubble_20') for n in ind_reviewtest]
    rate1 = [n.find_all(class_ = 'ui_bubble_rating bubble_10') for n in ind_reviewtest]

    rates_list = [rate1[i] + rate2[i] + rate3[i]+ rate4[i]+ rate5[i] for i in range(len(rate1))]

    rates = [item for sublist in rates_list for item in sublist]

    stay_date = str(soup.find_all(class_ = 'prw_rup prw_reviews_stay_date_hsx')).replace('<div class="prw_rup prw_reviews_stay_date_hsx" data-prwidget-init="" data-prwidget-name="reviews_stay_date_hsx"><span class="stay_date_label">Dato for besøg:</span> ', '').replace('</div>', '').replace('[', '').replace(']', '').replace(' ', '').split(',')
    
    rest_rates.append(rates)
    rest_rating_date.append(stay_date)
    
    end = time.time()
    timer = timer + end - start
    timer_accu = str(round(timer/60, 1))
    print("The loop has run for " + timer_accu + " minutes." )
```
