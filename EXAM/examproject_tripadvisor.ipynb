```python
from bs4 import BeautifulSoup
import requests
import re
import time
import scraping_class
import pandas as pd
import numpy as np
logfile = 'log.csv'
connector = scraping_class.Connector(logfile)
```

```python
# tripadvisor, get all url pages

numbers = [30*i for i in range(49)]
```

```python
# Tripadvisor stuff

list_of_rest_links_pre = []

for number in numbers:
    url = f'https://www.tripadvisor.dk/RestaurantSearch-g189541-oa{number}-a_date.2020__2D__08__2D__21-a_people.2-a_time.20%3A00%3A00-a_zur.2020__5F__08__5F__21-Copenhag.html#EATERY_LIST_CONTENTS'
    response = connector.get(url,'url_getter')
    html = response[0].text
    soup = BeautifulSoup(html,'html.parser')
    list_of_rest_links_pre.append([link['href'] for link in soup.find_all(class_ = '_15_ydu6b',href = True)])# = [link['href'] for link in soup.find_all(class_ = '_15_ydu6b',href = True)]

list_of_rest_links_01 = [item for sublist in list_of_rest_links_pre for item in sublist]

list_of_rest_links_01[:10]
```

```python
tripadvisor = 'https://www.tripadvisor.dk'

list_of_rest_links = [tripadvisor + i for i in list_of_rest_links_01]

list_of_rest_links[:10]
```

```python
rest_name = []
rest_location = []
rest_rating = []
rest_cuisine = []
rest_cuisine_2 = []
rest_cuisine_3 = []
rest_no_review = []
rest_price_cat = []

for link in list_of_rest_links:
    response3 = connector.get(link,'rest_data_getter')
    html3 = response3[0].text
    soup3 = BeautifulSoup(html3,'html.parser')
    try:
        rest_name1 = soup3.find(class_ = '_3a1XQ88S').get_text(strip=True)
    except:
        rest_name1 = ''
    try:
        rest_location1 = soup3.find(class_ = '_2saB_OSe').get_text(strip=True)
    except:
        rest_location1 = ''
    try:
        rest_rating1 = soup3.find(class_ = 'r2Cf69qf').get_text(strip=True).replace(",", ".")
    except:
        rest_rating1 = ''
    try:
        rest_cuisine1 = str(soup3.find_all(class_ = '_2mn01bsa')).split(",")[1].split('html">')[1].replace('</a>', '')
    except:
        rest_cuisine1 = ''
    try:
        rest_cuisine2 = str(soup3.find_all(class_ = '_2mn01bsa')).split(",")[2].split('html">')[1].replace('</a>', '')
    except:
        rest_cuisine2 = ''
    try:
        rest_cuisine3 = str(soup3.find_all(class_ = '_2mn01bsa')).split(",")[3].split('html">')[1].replace('</a>]', '')
    except:
        rest_cuisine3 = ''
    try:
        rest_no_review1 = soup3.find(class_ = '_10Iv7dOs').get_text(strip=True).split(" ")[0].replace('.', '')
    except:
        rest_no_review1 = ''
    try:
        rest_price_catdollar1 = soup3.find(class_ = '_2mn01bsa').get_text(strip=True).count('$')
    except:
        rest_price_catdollar1 = ''
    try:
        rest_price_catdash1 = soup3.find(class_ = '_2mn01bsa').get_text(strip=True).count('-')
    except:
        rest_price_catdash1 = ''
    try:
        rest_price_cat1 = rest_price_catdollar1 / (1 + rest_price_catdash1)
    except:
        rest_price_cat1 = ''
    rest_name.append(rest_name1)
    rest_location.append(rest_location1)
    rest_rating.append(rest_rating1)
    rest_cuisine.append(rest_cuisine1)
    rest_cuisine_2.append(rest_cuisine2)
    rest_cuisine_3.append(rest_cuisine3)
    rest_no_review.append(rest_no_review1)
    rest_price_cat.append(rest_price_cat1)
    print("Completed " + str(len(rest_name)) + " of 1807 links.")
    time.sleep(0.5)
```

```python
df = pd.DataFrame({'links': list_of_rest_links, 'rest_name' : rest_name, 'rest_location' : rest_location, 'rest_rating' : rest_rating, 'rest_cuisine' : rest_cuisine, 'rest_cuisine_2' : rest_cuisine_2, 'rest_cuisine_3' : rest_cuisine_3, 'rest_no_review' : rest_no_review, 'rest_price_cat' : rest_price_cat})

df2 = df.drop_duplicates(subset=['rest_name'], keep='first').copy()

df2
```

```python
df2['rest_price_cat'].replace('', np.nan, inplace=True)

df2['rest_price_cat'].replace(0.0, np.nan, inplace=True)

df2
```

```python
df2.dropna(subset=['rest_price_cat'], inplace=True)

df2
```
